```python
csv_file = "Don't tag as Auspost - Not Auspost.csv"
zip_df = pd.read_csv(csv_file, header=None)

zip_codes = zip_df[0].tolist()

url = "https://my-coffee-hunt.myshopify.com/admin/api/2024-01/orders.json?limit=200"
auth = ("a7208cd742d8594c25e538b7ee61d654", "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")

response = requests.get(url, auth=auth)
orders = response.json()['orders']

order_ids = []

for order in orders:
    zip_code = order['billing_address']['zip']
    try:
        if int(zip_code) not in zip_codes:
            print(zip_code,order['id'])
            order_ids.append(order['id'])
    except:
        continue
        
print(order_ids)
len(order_ids)
```

```python
for id in order_ids:
    url = f"https://my-coffee-hunt.myshopify.com/admin/api/2024-01/orders/{id}.json"
    auth = ("a7208cd742d8594c25e538b7ee61d654", "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")

    data = {
        "order":{
            "id":id,
            "tags":['Auspost']
        }
    }
    response = requests.put(url,json=data, auth=auth,headers={"Content-Type":"application/json"})
```
